```python
# import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x= 5*np.random.rand(150,1)-3
y=0.5 * x**2 + 1.5*x + 2 + np.random.rand(150,1)
#quadratic equation is y=0.5x^2 + 1.5x + 2
```

```python

plt.scatter(x,y)
plt.xlabel('x_data')
plt.ylabel('y_data')
plt.show()
```

```python
from sklearn.model_selection import  train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# implement simple linear regression

from sklearn.linear_model import LinearRegression

simple_linear_model = LinearRegression()
simple_linear_model.fit(x_train, y_train)
```

```python
from sklearn.metrics import  r2_score
score=r2_score(y_test,simple_linear_model.predict(x_test))
print('r2_score:',score)
```

```python
# visualize the results of the model
plt.plot(x_train,simple_linear_model.predict(x_train),color='blue')
plt.scatter(x_train, y_train, color='red')
plt.title('Linear Regression')
plt.xlabel('x data')
plt.ylabel('y')
plt.show()
```

```python
# Build a polynomial transformation
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=2, include_bias=True)
x_train_poly = poly.fit_transform(x_train)
x_test_poly = poly.transform(x_test)
```

```python
x_train_poly
```

```python
x_test_poly
```

```python
# Implement simple linear regression after performing polynomial transformation with degree 2
simple_linear = LinearRegression()
simple_linear.fit(x_train_poly, y_train)
score=r2_score(y_test,simple_linear.predict(x_test_poly))
print('r2_score:',score)
```

```python
print("coefficents:",simple_linear.coef_)
print("intercet:",simple_linear.intercept_)
```

```python
plt.scatter(x_train,simple_linear.predict(x_train_poly))
plt.scatter(x_train,y_train)
```

```python
# Perform polynomial transformation with degree 4
poly=PolynomialFeatures(degree=3,include_bias=True)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)

# Implement simple linear regression after performing polynomial transformation with degree 4
simple_linear = LinearRegression()
simple_linear.fit(x_train_poly, y_train)
score=r2_score(y_test,simple_linear.predict(x_test_poly))
print('r2_score:',score)
print("coefficents:",simple_linear.coef_)
print("intercet:",simple_linear.intercept_)
```

```python
plt.scatter(x_train,simple_linear.predict(x_train_poly))
plt.scatter(x_train,y_train)
```
